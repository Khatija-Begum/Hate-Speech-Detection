```python
import re      #regualr exp to convert simples of tweet to simple
import nltk   #natural lang toolkit  for removing stop words and stemming the words
import string #to remove punctuation
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords=set(stopwords.words('english'))
```

```python
stemmer=nltk.SnowballStemmer('english')
```

```python
#data cleaning

def clean_data(text):
  text=str(text).lower()
  text=re.sub('https://\S+|www\.S+',"",text)
  text=re.sub('\[.*?\]','',text)
  text=re.sub('<.*?>+','',text)
  text=re.sub('[%s]' %re.escape(string.punctuation),'',text)
  text=re.sub('\n','',text)
  text=re.sub('\w*\d\w*','',text)
  #stop word removal
  text=[word for word in text.split('  ') if word not in stopwords]
  text=''.join(text)
  #stemming the text
  text=[stemmer.stem(word) for word in text.split(' |')]
  text=''.join(text)
  return text
```

```python
from nltk.stem import PorterStemmer
import pandas as pd
```

```python
data['tweet']=data['tweet'].apply(clean_data)
```

```python
data
```

```python
x=np.array(data['tweet'])
y=np.array(data['labels'])
```

```python
x
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
cv=CountVectorizer()
x=cv.fit_transform(x)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=987)
```

```python
x_train
```

```python
#building out machine learning model

from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
#confusion matrix and accuracy

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.heatmap(cm,annot=True,fmt='f',cmap='YlGnBu')
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
